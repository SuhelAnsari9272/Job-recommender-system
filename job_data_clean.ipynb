```python
#importing required libraries
import pandas as pd
import numpy as np
```

```python
#reading the data
job_data = pd.read_csv(r'Careers_recommendation\Company Master_data.csv')
```

```python
#data overview
job_data.head()
```

```python
# we don't need some columns, we can remove them
#columns to remove
remove_cols = ['JD Approved','Jd Link Original','Stipend Range (in thousands)',
       'Domain Name', 'JD Punch Date', 'JD Last Registration Date','Recruiter', 'CR', 'Remarks','Type 1', 'Type 2',
       'Hr Email','#Reached Assignment',
       '#Reached Assignment: Awaited', '#Reached Assignment: Cleared',
       '#Reached Assignment: Not Appeared', '#Reached Assignment: Rejected',
       '#Reached Interview : Awaited', '#Reached Interview : Cleared',
       '#Reached Interview : Not Appeared', '#Reached Interview : Rejected',
       '#Reached Interview', '#Reached Other Rounds',
       'Company Shortlist - Awaited', 'Company Shortlist - Cleared',
       'Company Shortlist - No Action', 'Company Shortlist - Rejected',
       'Internal Shortlist : Rejected', 'Internal Shortlist: No Action',
       'Internal Shortlisted', 'JD Latest Release', 'Offered Learner Count']
```

```python
job_df = job_data.drop(remove_cols, axis= 1)
```

```python
job_df.head()
```

```python
job_df.head()
```

```python
# data info
job_df.info()
```

```python
# rename the columns
rename_col = {'job Id (va_job)' : 'job id', 'CTC Range (in lacs)' : 'CTC', 'Registrations - Interested' : 'Registrations',
              'Overall Exp Range' : 'Overall Exp', 'Relevant Exp Range' : 'Relevant Exp'}
job_df = job_df.rename(columns= rename_col)
```

```python
job_df.head()
```

```python
# impute the CTC with 0 if null
job_df.fillna({'CTC': 0 }, inplace = True)
```
